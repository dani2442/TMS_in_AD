```python
# Imports and directories
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pathlib import Path

SPINE = Path.cwd().parents[2]
DATA_DIR = SPINE / "data"
RES_DIR = SPINE / "results"
G_DIR = RES_DIR / "Finding_best_G"
AVG_DIR = G_DIR / "AvgHC"
INT_DIR = G_DIR / "Find_intercept"

# Load dictionary
fitting_dict = pd.read_pickle(AVG_DIR /"fitting_dict_synch.pkl")
wes = np.arange(0, 3., 0.01)
#Plot FC / phFCD to find the best G
plt.figure()
plt.plot(wes, fitting_dict['FC'], '-')
plt.plot(wes, fitting_dict['phFCD'], '-')
plt.annotate(f"Best phFCD at {wes[fitting_dict['phFCD'].argmin()]}", (1.35,0.9))
plt.xlim(0,2)
plt.xlabel("G")
plt.ylim(0,1.)
plt.ylabel("FC / phFCD")
plt.title("Finding the best G")
plt.savefig(AVG_DIR / "find_best_G.png")
```

```python
# Load results from fitting dictionary for different intercepts
fitting_dict_int = pd.read_pickle(INT_DIR /"fitting_dict_synch.pkl")
bs = np.arange(-0.01,0.01, 0.001)
# Plot intercepts results
plt.figure()
plt.plot(bs, fitting_dict_int['FC'], '-')
plt.plot(bs, fitting_dict_int['phFCD'], '-')
plt.annotate(f"Best phFCD at {round(bs[fitting_dict_int['phFCD'].argmin()], 3)}", (.0, 0.5))
plt.xlim(-0.01,0.01)
plt.xlabel("G")
plt.ylim(0,1.)
plt.ylabel("FC / phFCD")
plt.title("Finding the best G")
plt.savefig(INT_DIR / "find_best_G.png")
```
