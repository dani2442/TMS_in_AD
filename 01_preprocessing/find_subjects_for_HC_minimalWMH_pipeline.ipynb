```python
import numpy as np
import pandas as pd

root_dir = '/home/riccardo/ADNI_Hopf/Results/'
HC_names = np.load(root_dir + 'subject_list_wmh_volumes_HC.npy')
HC_WMH_volumes = np.load(root_dir + 'wmh_volumes_HC.npy')
HC_flag = ['HC' for i in HC_names]
MCI_names = np.load(root_dir + 'subject_list_wmh_volumes_MCI.npy')
MCI_WMH_volumes = np.load(root_dir + 'wmh_volumes_MCI.npy')
MCI_flag = ['MCI' for i in MCI_names]
```

```python
names_for_df = np.hstack([HC_names, MCI_names])
WMH_for_df = np.hstack([HC_WMH_volumes, MCI_WMH_volumes])
prepare_df = np.vstack([names_for_df, WMH_for_df]).T
flags_for_df = HC_flag + MCI_flag
```

```python
df_HC_minimalWMH = pd.DataFrame(prepare_df, columns=['Subjects', 'WMH_volumes'])
df_HC_minimalWMH['WMH_volumes'] = df_HC_minimalWMH['WMH_volumes'].astype(float)
df_HC_minimalWMH['Flags'] = flags_for_df
```

```python
subject_list_wmh_volumes_HC_minimal = df_HC_minimalWMH[(df_HC_minimalWMH['WMH_volumes'] <= 50) & (df_HC_minimalWMH['Flags'] == 'HC')]
subject_list_wmh_volumes_HC_other = df_HC_minimalWMH[(df_HC_minimalWMH['WMH_volumes'] > 50) & (df_HC_minimalWMH['Flags'] == 'HC')]
subject_list_wmh_volumes_MCI = df_HC_minimalWMH[df_HC_minimalWMH['Flags'] == 'MCI']
```

```python
np.save(root_dir + 'G_fitted_to_HC-minimalWMH/subject_list_wmh_volumes_HC_minimal', subject_list_wmh_volumes_HC_minimal.Subjects.to_numpy())
np.save(root_dir + 'G_fitted_to_HC-minimalWMH/subject_list_wmh_volumes_HC_other', subject_list_wmh_volumes_HC_other.Subjects.to_numpy())
np.save(root_dir + 'G_fitted_to_HC-minimalWMH/subject_list_wmh_volumes_MCI', subject_list_wmh_volumes_MCI.Subjects.to_numpy())



np.save(root_dir + 'G_fitted_to_HC-minimalWMH/wmh_volumes_HC_minimal', subject_list_wmh_volumes_HC_minimal.WMH_volumes.to_numpy())
np.save(root_dir + 'G_fitted_to_HC-minimalWMH/wmh_volumes_HC_other', subject_list_wmh_volumes_HC_other.WMH_volumes.to_numpy())
np.save(root_dir + 'G_fitted_to_HC-minimalWMH/wmh_volumes_MCI', subject_list_wmh_volumes_MCI.WMH_volumes.to_numpy())
```
